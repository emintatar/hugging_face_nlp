```python
model = AutoModel.from_pretrained("sentence-transformers/bert-base-nli-mean-tokens")
tokenizer = AutoTokenizer.from_pretrained("sentence-transformers/bert-base-nli-mean-tokens")
```

```python
def get_embedding(text):
    inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True)
    outputs = model(**inputs)
    embeddings = outputs.last_hidden_state.mean(dim=1)
    return embeddings

def calculate_similarity(text1, text2):
    embed1 = get_embedding(text1)
    embed2 = get_embedding(text2)
    cosine_sim = torch.nn.functional.cosine_similarity(embed1, embed2)
    return cosine_sim

text1 = "Insider is a global marketing technology company specializing in providing data-driven solutions to enhance customer engagement and optimize marketing strategies across various digital channels."
text2 =  "Insider offers advanced marketing technology solutions designed to improve customer interactions and maximize the effectiveness of digital marketing campaigns through data analytics and personalized strategies."

similarity = calculate_similarity(text1, text2)
print(f"Similarity: {similarity.item()}")
```
